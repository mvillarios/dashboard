```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_movies_meta = pd.read_csv('datasets/IMBD/movies_metadata.csv',
                             low_memory=False)

print("Archivo cargado correctamente")
```

```python
df_movies_meta = df_movies_meta[['id', 'title','vote_average', 'vote_count', 'budget', 'revenue']]

df_movies_meta.head()
```

```python
df_movies_meta = df_movies_meta.loc[df_movies_meta['vote_count'] >= 1500]

df_movies_meta.head()
```

```python
df_movies_meta.sort_values(by='vote_average', ascending=False, axis=0, inplace=True)

df_movies_meta.head(10)
```
